```python
# Estrutura Econômica e Mercado de Trabalho

# Tabela 1.8 rend UF - Rendimento médio real do trabalho principal habitualmente recebido por mês pelas pessoas 
# de 14 anos ou mais de idade, ocupadas na semana de referência, por sexo, cor ou raça e grupos de idade, 
# com indicação do coeficiente de variação, segundo Grandes Regiões, Unidades da Federação e Municípios das Capitais - 2019
file = '/home/mario/Documentos/LCDIA/Analytics/Datasets/Sintese Indicadores Sociais/csv/Tabela 1.8 rend UF.csv'
df_dados_ibge_tabela_01 = pd.read_csv(file)

# Tabela 1.8 rend UF - Rendimento médio real do trabalho principal habitualmente recebido por mês pelas pessoas 
# de 14 anos ou mais de idade, ocupadas na semana de referência, por sexo, cor ou raça e grupos de idade, 
# com indicação do coeficiente de variação, segundo Grandes Regiões, Unidades da Federação e Municípios das Capitais - 2019
file = '/home/mario/Documentos/LCDIA/Analytics/Datasets/Sintese Indicadores Sociais/csv/Tabela 1.13 rend UF.csv'
df_dados_ibge_tabela_02 = pd.read_csv(file)

# Tabela 1.24 UF - Taxa de subutilização composta da força de trabalho, por sexo, grupos de idade e cor ou raça, 
# com indicação do coeficiente de variação, segundo Grandes Regiões, Unidades da Federação e Municípios das Capitais - 2019
file = '/home/mario/Documentos/LCDIA/Analytics/Datasets/Sintese Indicadores Sociais/csv/Tabela 1.24 UF.csv'
df_dados_ibge_tabela_03 = pd.read_csv(file)


# Padrão de Vida e Distribuição de Renda

# Tabela 2.11 UF - Índice de Gini da distribuição do rendimento domiciliar per capita, segundo as Grandes Regiões, 
# as Unidades da Federação e os Municípios das Capitais
file = '/home/mario/Documentos/LCDIA/Analytics/Datasets/Sintese Indicadores Sociais/csv/Tabela 2.11 UF.csv'
df_dados_ibge_tabela_04 = pd.read_csv(file)

# Tabela 2.12 UF - Distribuição do rendimento domiciliar per capita, por classes de percentual de pessoas em ordem 
# crescente de rendimento, e Índice de Palma, com indicação do coeficiente de variação, segundo as Grandes Regiões, as Unidades da Federação e os Municípios das Capitais
file = '/home/mario/Documentos/LCDIA/Analytics/Datasets/Sintese Indicadores Sociais/csv/Tabela 2.12 UF.csv'
df_dados_ibge_tabela_05 = pd.read_csv(file)


# Educação

# Tabela 3.2 UF - Taxa ajustada de frequência escolar líquida das pessoas de 6 a 24 anos de idade, por grupos de 
# idade e nível de ensino, com indicação do coeficiente de variação, segundo as Grandes Regiões, as Unidades da Federação e os Municípios das Capitais
file = '/home/mario/Documentos/LCDIA/Analytics/Datasets/Sintese Indicadores Sociais/csv/Tabela 3.2 UF.csv'
df_dados_ibge_tabela_06 = pd.read_csv(file)

# Tabela 3.11 UF - Distribuição percentual de jovens de 15 a 29 anos de idade, por tipo de atividade na semana de 
# referência e grupos de idade, com indicação do coeficiente de variação, segundo Grandes Regiões, Unidades da Federação e Municípios das Capitais
file = '/home/mario/Documentos/LCDIA/Analytics/Datasets/Sintese Indicadores Sociais/csv/Tabela 3.11 UF.csv'
df_dados_ibge_tabela_07 = pd.read_csv(file)
```

```python
# Merges
df_dados_ibge = pd.merge(df_dados_ibge_tabela_01, df_dados_ibge_tabela_02, on='cidade', how='inner')
df_dados_ibge = pd.merge(df_dados_ibge, df_dados_ibge_tabela_03, on='cidade', how='inner')
df_dados_ibge = pd.merge(df_dados_ibge, df_dados_ibge_tabela_04, on='cidade', how='inner')
df_dados_ibge = pd.merge(df_dados_ibge, df_dados_ibge_tabela_05, on='cidade', how='inner')
df_dados_ibge = pd.merge(df_dados_ibge, df_dados_ibge_tabela_06, on='cidade', how='inner')
df_dados_ibge = pd.merge(df_dados_ibge, df_dados_ibge_tabela_07, on='cidade', how='inner')
```
